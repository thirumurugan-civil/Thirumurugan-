```python

import random

# Simulate sensor data (in reality, this would come from actual sensors)
def get_sensor_data():
    # Randomly generate position of an obstacle
    obstacle_position = random.randint(0, 100)
    return obstacle_position

# Perception module (simplistically processes sensor data)
def perceive_environment(sensor_data):
    if sensor_data > 50:
        print("Obstacle detected ahead!")
        return True
    return False

# Planning module (decides what to do based on perceptions)
def plan_route(obstacle_detected):
    if obstacle_detected:
        return "STOP"
    else:
        return "CONTINUE"

# Control module (executes the commands from planning)
def control_vehicle(command):
    if command == "STOP":
        print("Vehicle stopping.")
    else:
        print("Vehicle moving.")

# Main function to simulate autonomous vehicle operations
def autonomous_vehicle_system():
    sensor_data = get_sensor_data()
    obstacle_detected = perceive_environment(sensor_data)
    command = plan_route(obstacle_detected)
    control_vehicle(command)

# Run the system
autonomous_vehicle_system()
```
